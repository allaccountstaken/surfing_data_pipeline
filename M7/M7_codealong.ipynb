```python
nba_db = sqlite3.connect("nba.db")
```

```python
pd.read_sql_query('SELECT * FROM players_season LIMIT 1', nba_db).T
```

```python
mq = '''
SELECT * FROM players_season
WHERE LOWER(player) LIKE '%lebron%'

'''
pd.read_sql_query(mq, nba_db)
```

```python
pd.read_sql_query("SELECT * FROM players_season LIMIT 10", nba_db).T
```

```python
mq2 = '''
SELECT  player, season_pts + 1.5 * season_ast as mvp_metric
FROM players_season
ORDER BY mvp_metric DESC
LIMIT 10
'''
pd.read_sql_query(mq2, nba_db)
```
