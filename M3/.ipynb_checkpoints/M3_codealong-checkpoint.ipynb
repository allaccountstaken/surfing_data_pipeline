```python
import pandas as pd
import json
import requests
```

## First URL

```python
url1 = "https://www.thecocktaildb.com/api/json/v1/1/filter.php?c=Cocktail"
req1 = requests.get(url1, headers = {'User-agent':'Dima'})
req1
```

```python
my_json = json.loads(req1.text)
```

```python
my_json
```

```python
my_json['drinks'][:4]
```

```python
cocktail_df = pd.json_normalize(my_json, record_path=['drinks'])
cocktail_df.head()
```

```python
cocktail_df.idDrink.head()
```

## Second URL

```python
url2 = ('https://openlibrary.org/authors/OL23919A/works.json')
req2 = requests.get(url2)
other_json = json.loads(req2.text)
other_json
```

```python
df2 = pd.json_normalize(other_json, record_path=['entries'])
df2.head()
```

```python
other_json['entries'] # first step to get to the list level
```

```python
other_json['entries'][0] #second step - next authors
```

```python
other_json['entries'][0]['authors'] # At this lebvel iom a list again
```

```python
other_json['entries'][0]['authors'][0] # Pick at the first (zero) auther from the list
```

```python
other_json['entries'][0]['authors'][0]['author']['key'] # Finally, get the key
```

```python
# To get this key number for all the books, iterate over the first zero.
# This could be simpliefied with list comprehensions to get the first author:
[j['authors'][0]['author']['key'] for j in other_json['entries']][:10]
```

## Third URL

```python
url3 = 'https://en.wikipedia.org/w/api.php?action=query&list=search&srsearch=Virginia&format=json&srlimit=500'
req3 = requests.get(url3)
virginia_json = json.loads(req3.text)
virgina_json
```

```python
wikidata = pd.json_normalize(virginia_json, record_path=['query', 'search'])
wikidata.head(10)
```
