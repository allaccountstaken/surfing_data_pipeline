## Acquiring Data from APIs

```python
import numpy as np
import pandas as pd
import json
import requests
from datetime import datetime
```

```python
# Prepare URL
root = "http://aircasting.habitatmap.org"
endpoint = "/api/mobile/sessions.json"
measurment = "/api/measurements.json"
```

```python
# Date and time stamp
start = datetime(2021, 2, 9, 12, 0, 0).timestamp()
stop = datetime(2022, 2, 9, 12, 0, 0).timestamp()
```

```python
# Defining session parameters
sesion_params = {'west':-77.601242,
                'east':-77.352676,
                'south': 37.468319,
                'north': 37.639791,
                'time_from':start,
                'time_to':stop,
                 'tags':""
                }

# Check if params are well formated
json.dumps(sesion_params)
```

```python
# Check requests object status
r = requests.get(root + endpoint + measurment, params = {'q':json.dumps(sesion_params)})
r
```

```python
# Defining session params
session_parameters = {'west':-77.601242,
                     'east':-77.352676,
                     'south':37.468319,
                     'north':37.639791,
                     'time_from': start,
                     'time_to': stop,
                     'tags':"",
                     'usernames':"",
                     'limit': 1000,
                     'offset': 0,
                     'sensor_name': 'airbeam2-pm2.5',
                     'measurement_type': 'Particulate Matter',
                     'unit_symbol': 'µg/m³'}
```

```python
# Check reuests object status
r = requests.get(root + endpoint, params = {'q':json.dumps(session_parameters)})
r
```

```python
# Normalize by `sessions`
pd.json_normalize(json.loads(r.text), record_path=['sessions'])
```

```python
# Store sessions data into a dataframe, check the head:
session_df = pd.json_normalize(json.loads(r.text), record_path=['sessions'])
session_df.head()
```

```python
# Checkng just the `ids` column, to be used in `measurements` APIs, i.e. different endpoint.
session_df['id']
```

```python
# Define streams using stream `ids` from the sessions data
stream_ids = session_df['streams.AirBeam2-PM2.5.id']
```

```python
# Check requests object status
r = requests.get(root + measurment, 
                 params={'stream_ids':stream_ids[1]})
r
```

```python
# View the results in a tabel format:
pd.json_normalize(json.loads(r.text))
```
